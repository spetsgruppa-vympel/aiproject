```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

import pandas as pd

train_df = pd.read_csv("train.csv")
test_df  = pd.read_csv("test.csv")
test_with_incident_id = pd.read_csv("test.csv") # needed for submissions :3

drop_cols = ["Incident_ID", "Year", "Country"]

y_train = train_df["Attack Source"] # source
train_df = train_df.drop(columns=drop_cols + ["Attack Source"], errors="ignore") # attack source is the column we need to predict so down it goes
test_df.drop(columns=drop_cols, inplace=True, errors="ignore")
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```

```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# split columns
num_cols = train_df.select_dtypes(include=["number"]).columns
cat_cols = train_df.select_dtypes(exclude=["number"]).columns

# scale ts and stop the strings from being strings
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ]
)
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```

```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

# fit on train
x_train = preprocess.fit_transform(train_df)

# transform test x_x
x_test = preprocess.transform(test_df)

# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```
