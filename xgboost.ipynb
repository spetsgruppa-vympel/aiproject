```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```

```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

train_df = pd.read_csv("student_performance.csv")

drop_cols = ["Name", "Average_Score", "Math_Score", "Science_Score", "Language_Score", "History_Score"] # xgboost LOVES to overfit on LITERALLY EVERYTHING

y_train = train_df["Grade"]
train_df = train_df.drop(columns=drop_cols + ["Grade"], errors="ignore")

# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```

```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

cat_cols = train_df.select_dtypes(exclude=["number"]).columns

preprocess = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=True), cat_cols),
    ],
    remainder="passthrough"
)

# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```

```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

from sklearn.preprocessing import LabelEncoder

x_train = preprocess.fit_transform(train_df)

# xgboost needs numeric labels
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)

# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```

```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

xgb = XGBClassifier(
    n_estimators=1000,
    n_jobs=-1,
    max_depth=3,
    learning_rate=0.01,
    random_state=42,
    eval_metric="mlogloss"
)

scores = cross_val_score(xgb, x_train, y_train_enc, cv=4, scoring="accuracy")

print(f"cv scores: {scores}")
print(f"mean accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")

# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```

```python
# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM

from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    "max_depth":        [3, 4, 5, 6, 7, 8],
    "learning_rate":    [0.01, 0.05, 0.1, 0.2, 0.3],
    "n_estimators":     [100, 200, 300, 500, 1000],
    "subsample":        [0.6, 0.7, 0.8, 0.9],
    "colsample_bytree": [0.6, 0.7, 0.8, 0.9],
    "reg_alpha":        [0, 0.1, 1, 10],
    "reg_lambda":       [0.1, 1, 5, 10],
}

search = RandomizedSearchCV(
    XGBClassifier(n_jobs=-1, random_state=42, eval_metric="mlogloss"),
    param_distributions=param_dist,
    n_iter=100,
    cv=4,
    scoring="accuracy",
    random_state=42,
    verbose=1
)

search.fit(x_train, y_train_enc)

print(f"best accuracy: {search.best_score_:.4f}")
print(f"best params:   {search.best_params_}")

# In order to Know the RAM, you Must purchase The RAM. - Sun Tzu, Art of RAM
```
